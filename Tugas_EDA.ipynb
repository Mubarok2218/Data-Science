```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=df, x='Department', y='Productivity_Score') #Menggambar boxplot di mana sumbu X adalah 'Department' (kategori) dan sumbu Y adalah 'Productivity_Score' (numerik). Setiap kotak akan mewakili distribusi produktivitas untuk satu departemen.
plt.title('Produktivitas berdasarkan Departemen')
plt.xticks(rotation=45, ha='right')
#Faktor departemen tidak menjadi penentu utama produktivitas, karena distribusinya seragam.
```

```python
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Job_Level', y='Productivity_Score')
plt.title('Produktivitas berdasarkan Level Jabatan')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
#Produktivitas tidak meningkat signifikan seiring naiknya level jabatan.
```
